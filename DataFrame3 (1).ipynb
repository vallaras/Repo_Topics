# A Review of the .apply() Method on Single Columns

```python
info=pd.read_csv(r"C:\Users\ELCOT\Downloads\Finaltable.csv",index_col="Film")
info.sort_index(inplace=True)
info.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + " MILLIONS!"
info["Box Office"].apply(convert_to_string_and_add_millions)
```

```python
a=info["Box Office"].apply(convert_to_string_and_add_millions)
pd.DataFrame(a)
```

```python
info["Budget"]=info["Budget"].apply(convert_to_string_and_add_millions)
```

```python
columns=['Box Office','Budget','Bond Acotor Salary']
for col in columns:
    info[col] = info[col].apply(convert_to_string_and_add_millions)
```

```python
info.head(3)
```

# The .apply () Method with Row Values

```python
info=pd.read_csv(r"C:\Users\ELCOT\Downloads\Finaltable.csv",index_col="Film")
info.sort_index(inplace=True)
info.head(3)
```

```python
def good_movie(row):
    actor=row[1]
    budget=row[4]
    if actor=="Roger Moore":
        return "The best"
    elif actor=="Sean Connery" and budget > 40:
        return"Enjoable"
    else:
        return"i have no clue "
info.apply(good_movie,axis="columns")    
```
