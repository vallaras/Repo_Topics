# The Pivot Method

```python
sales=pd.read_csv(r"C:\Users\ELCOT\Downloads\salesman.csv",parse_dates=["Date"])
sales["Salesman"]=sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales["Salesman"]=sales["Salesman"].value_counts()
```

```python
pd.pivot_table(data=sales,index="Date",columns="Salesman",values="Revenue")
```
