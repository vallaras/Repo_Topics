```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**3
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Example')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(x,x**2,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,.8,.8])
axes.plot(x,y,'g')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('Example')
```

```python
fig = plt. figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'b')
axes1.set_xlabel('x_label_axes2')
axes1.set_ylabel('y_label_axes2')
axes1.set_title('axes1')

axes2.plot(x,y,'r')
axes2.set_xlabel('x_label_axes2')
axes2.set_ylabel('y_label_axes2')
axes2.set_title('axes2')
```
