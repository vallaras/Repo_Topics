```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = range(1,50)
y = [value * 3 for value in x]
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Draw line')
```

```python
X = [1,2,3]
Y = [2,4,1]
```

```python
plt.plot(X,Y,'b',marker='o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('example')
```

```python
x1 = [10,20,30]
y1 = [20,40,10]
```

```python
plt.plot(x1,y1,label ="line1")
x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2,y2,label="line2")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two or more lines')
```

```python
plt.plot(x1,y1, color='blue', linewidth = 3,  label = 'line1-width-3')
plt.plot(x2,y2, color='red', linewidth = 5,  label = 'line2-width-5')
plt.legend()
```

```python
plt.plot(x1,y1, color='blue', linewidth = 3,  label = 'line1-dotted',linestyle='dotted')
plt.plot(x2,y2, color='red', linewidth = 5,  label = 'line2-dashed', linestyle='dashed')
plt.title('example')
```

```python
x = [1,4,5,6,7]
y = [2,6,3,6,3]
```

```python
plt.plot(x,y,color='r',linestyle='dashdot',linewidth=3,marker='o',markerfacecolor='b',markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('dash-dot')
```
